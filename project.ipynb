#### Exercise 18. Normalize the variable of prices for each population and plot the 4 histograms in the same plot (you must use the subset obtained in Exercise 14) (★★★)

For the normalization method, you can use the one you consider; there is not a single correct answer to this question. Print the plot and write in the Markdown cell a brief analysis about the plot.

```python
# TODO
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')
cinturonsur=ds[ds["level5"].isin(["Fuenlabrada","Leganés", "Getafe", "Alcorcón"])]

#Creating a list to store normalized prices
normalized_prices = []


for city in ["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"]:
    #Getting indexes
    city_data = cinturonsur[cinturonsur["level5"] == city]
    
    #Getting price
    price = city_data["price"]
    
    #Applying the formula
    normalized_price = (price - price.min()) / (price.max() - price.min())
    normalized_prices.append(normalized_price)
    
    plt.hist(price, bins=20, alpha=0.5, label=city)

    

plt.legend()
plt.xlabel("Precio")
plt.ylabel("Frecuencia")
plt.title("Distribucion de precios en cinturón sur de Madrid")
plt.show()
```

#### Exercise 19. What can you say about the price per square meter (price/m2) between the towns of "Getafe" and "Alcorcón"? You must use the subset obtained in Exercise 14 (★★☆)

Hint: Create a new column called `pps` (price per square meter) and then analyze the values.

```python
# TODO
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd


ds = pd.read_csv('assets/real_estate.csv', sep=';')
cinturonsur = ds[ds["level5"].isin(["Fuenlabrada", "Leganés", "Getafe", "Alcorcón"])]
ds["pps"]=ds["price"]/ds["surface"]

valget = ds[ds["level5"] == "Getafe"]["pps"]
valalc = ds[ds["level5"] == "Alcorcón"]["pps"]

media_valget = valget.mean()
media_valalc = valalc.mean()

print(f"Precio medio del metro cuadrado en Getafe: {media_valget:.2f}")
print(f"Precio medio del metro cuadrado en Alcorcón: {media_valalc:.2f}")

```

Me resulta raro tras haber analizado la anterior gráfica que el precio en Alcorcón sea mas elevado que en Getafe, que es la población mas cara y que además, según se aprecia, dispone de más inmuebles. 

Dado que estamos analizando el precio por metro cuadrado, la conclusión lógica debería ser que los pisos en Alcorcón son más pequeños, pero no por ello más baratos. Esta sería la razón principal por la cual el precio medio es más elevado.
```python
## HERE: plot the coordinates of the estates

## PUT HERE YOUR CODE:
from ipyleaflet import Map, basemaps, Marker, AwesomeIcon
import pandas as pd

ds = pd.read_csv('assets/real_estate.csv', sep=';')
cinturonsur=ds[ds["level5"].isin(["Fuenlabrada","Leganés", "Getafe", "Alcorcón"])]

#Creating a dictionary to get latitudes and longitudes 
dic21 = cinturonsur.set_index('Unnamed: 0')[['latitude', 'longitude', 'level5']].to_dict(orient='index')


#Assigning colors to cities to implement AwesomeIcon
colores_ciudades = {
    'Alcorcón': 'orange',
    'Getafe': 'darkgreen',
    'Fuenlabrada': 'red',
    'Leganés': 'beige'
}

#Latitude/longitude converting to proper format ,/.
for nim, valores in dic21.items():
    latitude = float(valores['latitude'].replace(',', '.'))
    longitude = float(valores['longitude'].replace(',', '.'))
    level5 = valores['level5']

#Setting AwesomeIcon parameters
    icono = AwesomeIcon(name='home',  marker_color=colores_ciudades.get(level5, 'blue')) 
    
#Assigning markers by city.
    if level5 == 'Alcorcón':
        marcador = Marker(location=(latitude, longitude), icon=icono , draggable=True)
        map.add_layer(marcador)
    if level5 == 'Getafe':
        marcador = Marker(location=(latitude, longitude),icon=icono , draggable=True)
        map.add_layer(marcador)
    if level5 == 'Fuenlabrada':
        marcador = Marker(location=(latitude, longitude),icon=icono , draggable=True)
        map.add_layer(marcador)
    if level5 == 'Leganés':
        marcador = Marker(location=(latitude, longitude),icon=icono , draggable=True)
        map.add_layer(marcador)

map
```
